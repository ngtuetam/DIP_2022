```python
import cv2
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import matplotlib.image as mpimg
from skimage.feature import hog
from PIL import Image
import imageio
```

```python
def plot_hist(img):
    h = np.zeros((256))
    rows,cols,d = img.shape
    for i in range(rows):
        for j in range(cols):
            h[img[i,j]] = h[img[i,j]] +1
    return h
```

```python
img = cv2.imread('XLA-cuoiky-2022/final1.bmp')
Intensity = np.arange(0,256)
plt.plot(Intensity,plot_hist(img))
```
