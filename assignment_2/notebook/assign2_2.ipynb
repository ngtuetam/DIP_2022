### 2. Write your program to produce the eight bit-plane images of the 2_2.bmp image.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_path = '/home/ngtuetam/workspace/DIP_2022/assignment_2/images/2_2.bmp'
```

```python
img = plt.imread(img_path)
plt.imshow(img)
```

```python
img.max(), img.min()
```

```python
img_double = np.double(img)
```

```python
# extract each bit planes of original image to make 8 binary image
bp1 = np.mod(img_double, 2);
bp2 = np.mod(np.floor(img_double/2), 2);
bp3 = np.mod(np.floor(img_double/4), 2);
bp4 = np.mod(np.floor(img_double/8), 2);
bp5 = np.mod(np.floor(img_double/16), 2);
bp6 = np.mod(np.floor(img_double/32), 2);
bp7 = np.mod(np.floor(img_double/64), 2);
bp8 = np.mod(np.floor(img_double/128), 2);
```

```python
plt.rcParams['figure.figsize'] = [20, 20]
plt.subplot(2, 4, 1);
plt.imshow(bp1);
plt.title('Bit Plane 1 (LSB)');
plt.subplot(2, 4, 2);
plt.imshow(bp2);
plt.title('Bit Plane 2');
plt.subplot(2, 4, 3);
plt.imshow(bp3);
plt.title('Bit Plane 3');
plt.subplot(2, 4, 4);
plt.imshow(bp4);
plt.title('Bit Plane 4');
plt.subplot(2, 4, 5);
plt.imshow(bp5);
plt.title('Bit Plane 5');
plt.subplot(2, 4, 6);
plt.imshow(bp6);
plt.title('Bit Plane 6');
plt.subplot(2, 4, 7);
plt.imshow(bp7);
plt.title('Bit Plane 7');
plt.subplot(2, 4, 8);
plt.imshow(bp8);
plt.title('Bit Plane 8 (MSB)');
```
