```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
img_filepath = '/home/ngtuetam/workspace/DIP_2022/assignment_1/data/1_2.tif'
# plt.imshow(plt.imread(img_filepath))
```

```python
def convert_to_gray(image):
    """ Convert imgae to grayscale
    Args:
        image (np.array): input image as numpy array
    Returns:
        img_gray (np.array): gray image
    """
    
    img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return img_gray
```

```python
def rotate_image(image):
    """ Rotate image 45 degrees clockwise
    Args:
        image(np.array): input image as numpy array
    Returns:
        rotated_image(np.array): rotated image
    """
    
    rotated_image = image.rotate(45)
    return rotated_image
    
```

```python
img_np = plt.imread(img_filepath)
gray_image = convert_to_gray(image=img_np)
img_ = Image.open(img_filepath)
rotated_img = rotate_image(img_)

plt.subplot(1,3, 1)
plt.imshow(img_np)
plt.subplot(1,3, 2)
plt.imshow(gray_image, cmap='gray')
plt.subplot(1,3,3)
plt.imshow(rotated_img)
plt.show()
```
