```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from PIL import Image

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
img_filepath = '/home/ngtuetam/workspace/DIP_2022/assignment_1/data/1_4.bmp'
image = plt.imread(img_filepath)
plt.imshow(image)
```

```python
"""
Minimum and maximum data value for this image
"""
print(f"Maximun value of image: {image.min()}")
print(f"Minimun value of image: {image.max()}")
```

```python
"""
Convert to double 
"""
img_converted = image.astype(np.double)
plt.imshow(img_converted)
plt.show()
```

```python
"""
Correct the converted image
"""
img_converted_corrected = img_converted / 255
plt.imshow(img_converted_corrected)
plt.show()
```
