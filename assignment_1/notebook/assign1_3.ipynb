```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
img_filepath = '/home/ngtuetam/workspace/DIP_2022/assignment_1/data/1_2.tif'
#plt.imshow(plt.imread(img_filepath))
```

```python
def manipulate_image(image, grid_num):
    """ Manipulate given image and return the result
    
    Args:
        image (np.array): input image as numpy array
        grid_num (int): number of grids 
        
    Returns:
        _image (np.array): manipulated image
        
    Raises:
        None.
    
    """
    h, w = image.shape[:2]
    xsize = int(h/grid_num) 
    ysize = int(w/grid_num) 

    # create image Y with shape X
    _image = np.zeros_like(image)
    for i in range(grid_num):
        for j in range(grid_num):
            _image[
                xsize*(grid_num-i-1):xsize*(grid_num-i), 
                ysize*(grid_num-j-1):ysize*(grid_num-j)
            ] = image[xsize*(i):xsize*(i+1), ysize*(j):ysize*(j+1)]; 


    return _image
```

```python
img_np = plt.imread(img_filepath)
manipulated_img_np = manipulate_image(image=img_np, grid_num=4)


plt.subplot(1,2, 1)
plt.imshow(img_np)
plt.subplot(1,2,2)
plt.imshow(manipulated_img_np)
```
